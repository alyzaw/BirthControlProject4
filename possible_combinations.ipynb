```python
import os
import numpy as np
import itertools
```

```python
age_a = list(range(15, 45))
marstat = [1,2,3,4,5,6,9]
hispanic = [1,2]
race = [1,2,3]
hieduc = list(range(5, 15))
```

```python
#AGE_A	MARSTAT	RACE	HISP	NUMCHILD	HIEDUC
a = [age_a, marstat, race, hispanic, hieduc]
combinations = list(itertools.product(*a))
combinations
```

```python
import tensorflow as tf 
import os

import pandas as pd
import numpy as np

model = tf.keras.models.load_model("birth_control5features.h5")
model_output = []
num_combo = len(combinations)
i = 0
for c in combinations:
    i = i + 1
    input_dict = {'AGE_A': c[0], 'MARSTAT': c[1], 'RACE': c[2], 'HISPANIC': c[3], 'HIEDUC': c[4]}
    input_pd = pd.DataFrame(data = input_dict, index=[0])
    model_out = model.predict_classes(input_pd)
    model_output.append(model_out[0])
    print(f'Creating {i} of {num_combo}')

```

```python
combo_pd = pd.DataFrame({'Input': combinations, 'Output': model_output})
combo_pd.head()
```

```python
#combo_pd["Input"] = combo_pd["Input"].map(lambda x: x.lstrip('()'))
output = combo_pd.loc[combo_pd['Input'] == (15, 1, 1, 5)] 
output.iloc[0,1]
#combo_pd.head()
```
