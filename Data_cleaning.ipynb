```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame

all_data3 = pd.read_csv('2013_2015_FemRespData.csv', low_memory=False)
all_data2 = pd.read_csv('2011_2013_FemRespData.csv', low_memory=False)
all_data1 = pd.read_csv('2006_2010_FemResp.csv', low_memory=False)
```

```python
all_data1.columns
```

```python
cut_data3 = all_data3[['CASEID', 'CONSTAT1', 'RHADSEX' ,'AGE_A', 'MARSTAT', 'RACE', 'HISPANIC', 'HIEDUC', 'TOTINCR', 'LIFEPRT']]
cut_data2 = all_data2[['CASEID', 'CONSTAT1', 'RHADSEX' ,'AGE_A', 'MARSTAT', 'RACE', 'HISPANIC', 'HIEDUC', 'TOTINCR', 'LIFEPRT']]
cut_data1 = all_data1[['CASEID', 'CONSTAT1', 'RHADSEX' ,'AGE_A', 'MARSTAT', 'RACE', 'HISPANIC', 'HIEDUC', 'TOTINCR', 'LIFEPRT']]
```

```python
#merge 3 datasets

merged_data1 = cut_data3.merge(cut_data2.merge(cut_data1, how='outer'), how='outer')
merged_data1.head()
```

```python
len(merged_data1.index)
```

```python
#Remove people that had never had heterosexual sex
merged_data1 = merged_data1[merged_data1.RHADSEX !=5]
#remove people who are not currently on birth control
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 40]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 41]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 42]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 30]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 31]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 32]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 33]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 34]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 35]
merged_data1 = merged_data1[merged_data1.CONSTAT1 != 36]

#removed refused/don't know

merged_data1 = merged_data1[merged_data1.LIFEPRT != '999']
merged_data1 = merged_data1[merged_data1.LIFEPRT != '998']
```

```python
#create categories
# c_sterilization = [1, 2]
# c_norplantIUDDia  = [3, 10, 11]
# c_depo = [5]
# c_pillringpatch = [6, 7, 8]
# c_non_steroid = [12, 14, 17, 18]
# c_withdrawalcal = [19, 20, 21, 22]
bins_constat = [0,2,4,8,11,18,22]
labels_constat = [0,1,2,3,4,5]
bins_lifepart = []
labels_lifepart = []
merged_data1["consolidatedconst1"] = pd.cut(x=merged_data1['CONSTAT1'], bins=bins_constat, labels=labels_constat)
merged_data1.head(300)
```

```python
#merged_data1.loc[merged_data1['consolidatedconst1'] < -1]
merged_data1['consolidatedconst1'] = merged_data1['consolidatedconst1'].map({0: 0, 1: 1, 2: 2, 3: 3, 4: 1, 5: 5})
merged_data1.info()
```

```python
merged_data1.drop(columns = ['RHADSEX', 'CASEID', 'CONSTAT1', 'LIFEPRT', 'TOTINCR'], inplace = True)
```

```python
merged_data1 = merged_data1.dropna()
```

```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
data = merged_data1
```

```python
X = data.drop("consolidatedconst1", axis=1)
y = data["consolidatedconst1"]
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create a Neural Network model here
# YOUR CODE HERE
model = Sequential()
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train = X_train.reset_index(drop=True)
X_train
```

```python
X_train.iloc[2]
```

```python
model.add(Dense(units=100, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=9915, activation ="softmax"))
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["sparse_categorical_accuracy"])
```

```python
model.fit(
    X_train
    ,y_train
    ,epochs=100
    ,verbose=2)
```

```python
model.save("birth_control5features.h5")
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test, y_test, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
